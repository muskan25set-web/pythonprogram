Program of reversed digit and to check if it is palindrome or not.

```python
num=int(input("enter a number:"))
original = num #Store original value
reverse=0
while num>0: #Loop until all digits are processed
    digit=num%10 #Extract the last digit
    reverse=reverse*10+digit #Build the reversed number
    num=num//10 #Remove the last digit

if original==reverse: #Check if reversed number matches the original one
    print("Palindrome number")
else:
    print("Not an palindrome number")

```

Program to check if the number is armstrong or not.

```python
num=int(input())
original = num #Save original value to compare at the end
sum=0
while num>0:
    digit=num%10
    sum+=digit**3 #Add the cubes of digit to sum
    num=num//10 #Remove the last digit
if sum==original: #Compare the calculated sum with the original input
    print("Armstrong number")
else:
    print("Not an armstrong number")

```
