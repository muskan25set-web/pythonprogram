Program to print number 1 to 5.

```python
for i in range(1, 6): #Print numbers 1 through 5
    print(i)
```

Program to loop through a list.

```python
fruits = ["apple", "banana", "mango","cheery"] #List of fruit strings

for fruit in fruits: #Print each item in the list
    print(fruit)
```

Program using range() variation.

```python
for i in range(5): #Loop 5 times, strating from 0
    print(i)

for i in range(2, 11, 2): #Loop from 2 to 10, skipping by 2
    print(i)
```

Progam to print the given number's multiplication table.

```python
num = int(input("Enter a number: "))  #Take number input from user

for i in range(1, 11): #Take number input from user
    print(num, "x", i, "=", num * i)
```

To check the given number's factorial.

```python
num = int(input("Enter a number: "))
factorial = 1 #Initialize factorial variable

for i in range(1, num + 1):
    factorial *= i #Multiply each number to factorial

print("Factorial:", factorial)
```
