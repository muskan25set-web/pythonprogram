To check if student is pass or fail.

```python
a=int(input("Enter student marks:")) #Get marks from user
if a>=90:
    print("Grade A") #Condition
elif a>=80:
    print("Grade B")
elif a>70:
    print("Grade C")
elif a>60:
    print("PASS")
else:
    print("Fail") #Default case for marks < 60
```

To check weather the number is postive or negative

```python
a=int(input("enter a number:"))
if a>0: #Check if greater than 0
    print("The number is positive:")
elif a<0: #Check if less than 0
    print("The number is negative")
else: #If neither, it must be 0
    print("The number is zero")
```

To check the greater number.

```python
a=int(input("enter first number"))
b=int(input("enter second number"))
if a/b>1: #Use division ratio to compare
    print("a is greater")
elif a/b==1: #Ratio of 1 means numbers are identical
    print("a and b are equal")
else: #Ratio < 1 means b is larger
    print("b is greater")
```
